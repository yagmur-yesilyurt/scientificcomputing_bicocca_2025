```python
import numpy as np
import matplotlib.pyplot as plt
```

## Q2: Planetary positions

```python
a = np.array([0.39, 0.72, 1.00, 1.52, 5.20, 9.54, 19.22, 30.06, 39.48])
P = np.array([0.24, 0.62, 1.00, 1.88, 11.86, 29.46, 84.01, 164.8, 248.09])
names = ["Mercury", "Venus", "Earth", "Mars", "Jupiter", "Saturn",
         "Uranus", "Neptune", "Pluto"]

fig, ax = plt.subplots(figsize=(7, 5))

ax.loglog(a, P, "o", color="steelblue", markersize=6)

# label each planet -- offset a bit so text doesn't overlap the dot
for name, x, y in zip(names, a, P):
    ax.annotate(name, (x, y), textcoords="offset points", xytext=(6, 3), fontsize=8)

ax.set_xlabel("Semi-major axis (AU)")
ax.set_ylabel("Orbital period (years)")
ax.set_title("Planetary periods vs distances (log-log)")
ax.grid(True, which="both", linestyle="--", alpha=0.4)

plt.tight_layout()
plt.savefig("L03_Q2_planets.png", dpi=120)
plt.show()
```

The straight line on a log-log plot confirms Kepler's third law: $P \propto a^{3/2}$. Nice.

## Q3: A perfect circle

```python
def draw_circle(x0, y0, R, color, ax):
    """draw a filled circle centered at (x0, y0) with radius R"""
    theta = np.linspace(0, 2*np.pi, 300)
    x = x0 + R * np.cos(theta)
    y = y0 + R * np.sin(theta)
    ax.fill(x, y, color=color, alpha=0.6)

# draw 10 random circles
rng = np.random.default_rng(7)

fig, ax = plt.subplots(figsize=(6, 6))

colors = ["crimson", "steelblue", "seagreen", "darkorange", "mediumpurple",
          "gold", "tomato", "teal", "slategray", "hotpink"]

for c in colors:
    x0 = rng.uniform(-4, 4)
    y0 = rng.uniform(-4, 4)
    R  = rng.uniform(0.3, 1.5)
    draw_circle(x0, y0, R, c, ax)

ax.set_aspect("equal")  # important! otherwise circles look like ovals
ax.set_title("10 random circles")
ax.set_xlim(-6, 6)
ax.set_ylim(-6, 6)

plt.tight_layout()
plt.savefig("L03_Q3_circles.png", dpi=120)
plt.show()
```
