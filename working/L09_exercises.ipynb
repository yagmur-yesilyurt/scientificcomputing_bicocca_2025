```python
import sympy as sym

x = sym.Symbol('x')
sym.init_printing()
```

## Q1: Play with expressions

```python
f = x * sym.exp(-x) + x * (1 - x)
print("f =", f)
print()
```

```python
# evaluate at the given points
# .evalf() gives a decimal instead of leaving it as a sympy expression
for val in [0, 0.1, 0.2, 0.4, 0.8]:
    result = f.subs(x, val).evalf()
    print(f"f({val}) = {result:.6f}")
```

## Q2: Polynomials

```python
p = x**4 - 6*x**3 + x**2 + 24*x + 16

factored = sym.factor(p)
print("factored form:", factored)
```

```python
# find the zeros
zeros = sym.solve(p, x)
print("zeros:", zeros)
```

```python
# double-check: plug zeros back in, should all give 0
for z in zeros:
    print(f"  p({z}) = {p.subs(x, z).simplify()}")
```

## Q3: There and back again

```python
f = sym.sin(x) * sym.exp(-x)
print("original f:", f)
```

```python
# integrate f with respect to x
F = sym.integrate(f, x)
print("integral F:", F)
```

```python
# differentiate the result -- should get back f
dF = sym.diff(F, x)
print("dF/dx:     ", dF)

# simplify the difference to confirm it's zero
diff = sym.simplify(dF - f)
print("dF/dx - f =", diff)  # should be 0
```
