## Q1: Fun with arrays

```python
# Part A
# arange gives 1..15, reshape into 3 columns, then transpose to get the right shape
arr = np.arange(1, 16).reshape(3, 5).T
print(arr)
```

```python
# grab 2nd and 4th rows (index 1 and 3)
subset = arr[[1, 3], :]
print(subset)
```

```python
# Part B
# make a 6x6 array of zeros, then set the border to 1
# using slices so it works for any size
n, m = 6, 6
border = np.zeros((n, m), dtype=int)

border[0, :]  = 1   # top row
border[-1, :] = 1   # bottom row
border[:, 0]  = 1   # left column
border[:, -1] = 1   # right column

print(border)
```

## Q3: Are you faster than numpy? No

```python
def my_std(a):
    """compute standard deviation of a 1D array from scratch"""
    mean = np.sum(a) / len(a)
    variance = np.sum((a - mean)**2) / len(a)
    return np.sqrt(variance)

# test on a random array
rng = np.random.default_rng(0)
data = rng.normal(size=10000)

my_result = my_std(data)
np_result = np.std(data)

print(f"my std:    {my_result:.8f}")
print(f"numpy std: {np_result:.8f}")
print(f"difference: {abs(my_result - np_result):.2e}")
```

```python
# compare runtime -- spoiler: numpy wins
import timeit

t_mine = timeit.timeit(lambda: my_std(data), number=500)
t_np   = timeit.timeit(lambda: np.std(data), number=500)

print(f"my function: {t_mine:.3f} s")
print(f"numpy:       {t_np:.3f} s")
print(f"numpy is ~{t_mine/t_np:.1f}x faster")
```
