## Q1: Machine Precision

```python
# start with eps = 1 and keep halving until 1+eps == 1
eps = 1.0

while 1.0 + eps != 1.0:
    eps /= 2

# we went one step too far, so multiply back
eps *= 2

print("machine epsilon:", eps)
```

The result is around 2.22e-16, which matches what numpy reports as `np.finfo(float).eps`. Makes sense since Python uses 64-bit floats by default.

## Q2: Iterations

```python
# Part 1 - zip two lists together
names = ["Mary", "John", "Sarah"]
age = [21, 56, 98]

for name, a in zip(names, age):
    print(f"{name} is {a} years old")
```

```python
# Part 2 - enumerate with random numbers
import random

random.seed(42)  # so results are reproducible

values = [random.randint(0, 100) for _ in range(5)]

for i, val in enumerate(values):
    print(f"index {i} -> {val}")
```
